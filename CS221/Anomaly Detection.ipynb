### Gaussina (Normal) Distribution
Say $x \in \mathbb{R}$. If $x$ is a distrubuted Gaussian with mean $\mu$, variance $\sigma^2$, lwhere $\sigma$ is __standart derviation__, THUS: $x \text{~} N(\mu,\sigma^2)$, where "~" (=distrubted as) and "p(...)" is the probaliblity<br>

$$p(x; \mu, \sigma^2) = \frac{1}{\sqrt{2\pi}\sigma} exp(-\frac{(x-\mu)^2}{2 \sigma^2})$$
where: __NOTE:__ at sigma m can be alternate by $m-1$<br>

$$\mu = \frac{1}{m}\sum_\limits{i=1}^m x^{(i)} \text{ AND }\sigma^2 = \frac{1}{m }\sum_\limits{i=1}(x^{(i)}-\mu)^2$$
<br>
For all data of training set:
<br>$$\Pi_{j=1}^n p(x_j; \mu_j, \sigma_j^2$$

## Anomaly Detection Algortihm
1. Choose feateures $x_i$ that you could be considered as indicative of anomalous examples.<br>
2. Fit parameters $\mu_1,...,\mu_n, \sigma_1^2,...,\sigma_n^2$<br>

$$\mu_j = \frac{1}{m}\sum_\limits{i=1}^m x_j^{(i)}$$
$$\sigma_j^2 = \frac{1}{m}\sum_\limits{i=1}^m (x_j^{(i)}-\mu_j)^2$$
$$ \text{ it can be vectorized as follow i.e:} \mu = \text{ vector of } \mu_1 \text{through} \mu_n = \frac{1}{m}\sum_\limits{i=1}^m x^{(i)}$$

3. Given new example $x$, compute $p(x)$:
$$p(x) = \Pi_{j=1}^n p(x_j; \mu_j, \sigma_j^2 = \Pi_{j=i}^n \frac{1}{\sqrt{2\pi} \sigma_j} exp(-\frac{(x_j - \mu_j)^2}{2\sigma_j^2})$$
$$\therefore \text{ Anomaly if } p(x) < \epsilon$$

## Error Analysis for Anomaly Detection
Want $p(x)$ large for normal examples $x$ AND $p(x)$ small for anomalous examples $x$<br>

## Multivariate Gaussian (Normal) Distribution
### Intuition
<img src="pic\multivariant_gaussian_distribution.png"><br>
$$\therfore Mutlivearte Gaussian Distribution$$
$x \in \mathbb{R}^n$. Do not model $p(x_1),p(x_2),...,$ etc. separtely.<br>
Model $p(x)$ all in one go.<br>
Parameters: $\mu \in \mathbb{R}^n, \Sigma \in \mathbb{R}^{n\times n}$ (covariance matrix)<br>
$$p(x; \mu, \Sigma) = \frac{1}{(2\pi)^{n/2 \color{pink}{|\Sigma|^{1/2}}} exp(-\frac{1}{2}(x-\mu)^T \Sigma^{-1}(x-\mu)}$$
where, $|\Sigma|$ = determinat of $\Sigma$; compute by: det(Sigma)<br>
